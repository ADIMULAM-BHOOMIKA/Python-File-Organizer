```python
import os
import shutil
try:
    file_types = {
        ".txt": "TextFiles",
        ".csv": "CSV_Files",
        ".jpg": "Images",
        ".pdf": "PDFs"
    }

    for folder in set(file_types.values()):
        os.makedirs(folder, exist_ok=True)

    files = os.listdir()

    for file in files:

        if os.path.isfile(file):

            extension = os.path.splitext(file)[1]

            if extension in file_types:

                destination_folder = file_types[extension]

                shutil.move(file, f"{destination_folder}/{file}")

                print(f"{file} moved to {destination_folder}")

    with open("log.txt", "w") as log:
        log.write("Files organized successfully")

    with open("log.txt", "r") as log:
        print("\nLog File:")
        print(log.read())

except Exception as e:
    print("Error:", e)
```
